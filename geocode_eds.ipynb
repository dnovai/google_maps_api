```python
import googlemaps
import pandas as pd
```

```python
def get_geometry(direccion, comuna, region):

    lat = None
    lng = None

    address = direccion + ', '+ comuna + ', ' + region
    gcode_result = gmaps.geocode(address)
    if gcode_result:
        lat, lng = gcode_result[0]['geometry']['location']['lat'], gcode_result[0]['geometry']['location']['lng']
    return lat, lng
```

```python
eds_con_tienda_df = pd.read_csv('eds_con_tienda.csv', sep=';')
eds_con_tienda_df
```

```python
geocode_con_tienda = list(
    eds_con_tienda_df.apply(
        lambda x: get_geometry(
            direccion=x['direccion'],
            comuna=x['comuna'],
            region=x['region']) if (x['Latitud'] == 0.0 and x['Longitud'] == 0.0) else (x['Latitud']/(10**13), x['Longitud']/(10**13)), axis=1))

eds_con_tienda_df[['nueva_latitud', 'nueva_longitud']] = pd.DataFrame(geocode_con_tienda, index=eds_con_tienda_df.index)
eds_con_tienda_df
```
